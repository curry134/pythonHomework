```python
# Problem 4 code here (can split into multiple code chunks if you want)

LRschool= smf.ols(formula = 'Average_Class_Size ~ Average_Salary + Perc_Economically_Disadvantaged + Perc_English_Language_Learner', data = schoolData).fit()

LRschool.summary()

# R squared is roughly 6 percent, which means 6% of the models variance can be explained by the variables. The null hypothesis is rejected for the perc econ and perc english, which means there is evidence supporting there is a relationship between these predictor values and the class size.
```

```python
LRschool= smf.ols(formula = 'Average_Class_Size ~ Average_Salary + Perc_Economically_Disadvantaged + Perc_English_Language_Learner + School_Type', data = schoolData).fit()

LRschool.summary()
# R squared is roughly 6 percent, which means 6% of the models variance can be explained by the variables. The null hypothesis is rejected for the perc econ and perc english, which means there is evidence supporting there is a relationship between these predictor values and the class size. Additionally, it appears that there is a relationship between school type and average class size, as we reject the null hypothesis. 
```
